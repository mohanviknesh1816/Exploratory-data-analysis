```python
import pandas as pd
# Load the dataset
df = pd.read_csv(r"C:\Users\91739\Downloads\archive (2)\BankChurners.csv")
df.head()
```

```python
df.shape
```

```python
# List numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
print("Numeric Columns:", numeric_cols)
```

```python
df[numeric_cols].describe().T

```

```python
# Median
median_values = df[numeric_cols].median()
# Mode (use first mode value only)
mode_values = df[numeric_cols].mode().iloc[0]
# Combine 
summary = df[numeric_cols].describe().T
summary['median'] = median_values
summary['mode'] = mode_values

# Round for neatness
summary = summary.round(2)
summary
```

```python
# Selected features
selected_cols = ['Total_Trans_Amt', 'Total_Trans_Ct', 'Credit_Limit', 'Avg_Utilization_Ratio']

# Compute summary statistics
summary = df[selected_cols].describe().T

# Add median and mode
summary['median'] = df[selected_cols].median()
summary['mode'] = df[selected_cols].mode().iloc[0]

# Round the result
summary = summary.round(2)

# Display
summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
#histogram
plt.figure(figsize=(12, 8))
for i, col in enumerate(selected_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#boxplot
plt.figure(figsize=(12, 6))
for i, col in enumerate(selected_cols, 1):
    plt.subplot(2, 2, i)
    sn.boxplot(x=df[col], color='salmon')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```

```python
#  correlation
corr = df[selected_cols].corr()
#  heatmap
plt.figure(figsize=(6, 5))
sn.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
#pairplot
sn.pairplot(df[selected_cols], diag_kind='kde', corner=True)
plt.suptitle("Pairplot", y=1.02)
plt.show()
```

```python
#Distribution & Outlier Detection
plt.figure(figsize=(12, 6))
for i, col in enumerate(selected_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
#relationship between Total_Trans_Ct vs. Total_Trans_Amt
sns.scatterplot(data=df, x='Total_Trans_Ct', y='Total_Trans_Amt', alpha=0.5)
plt.title('Transaction Count vs Transaction Amount')
plt.show()
```

```python
#relationship between avg_Utilization_Ratio vs. Credit_Limit
sns.scatterplot(data=df, x='Credit_Limit', y='Avg_Utilization_Ratio', alpha=0.5)
plt.title('Credit Limit vs Avg Utilization Ratio')
plt.show()
```
